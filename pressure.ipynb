```python
import os
path = "/Users/fangzhihao/Desktop/港大学习/sem2/project/FacialExpressionRecognition-master/input/all"
files = os.listdir(path)
print(len(files))

```

```python
from recognition import *
from model import CNN3
import numpy as np
def NormalizeData(data):
    return (data - np.min(data)) / (np.max(data) - np.min(data))
model = CNN3()
model.load_weights('/Users/fangzhihao/Desktop/港大学习/sem2/project/FacialExpressionRecognition-master/models/cnn3_best_weights.h5')
train_data = []
train_labels = []
aaa = 0
for file_name in files:
    if not file_name.endswith(".jpg"):
        continue
    pressure_level = int(file_name[0])
    file_name = os.path.join(path, file_name)
    try:
        emotion, possibility = predict_expression(file_name, model)
    except:
        # print("error")
        continue
#     # normalize the possibility
    possibility = NormalizeData(possibility)
    if emotion != 'no':
        aaa += 1
        train_data.append(possibility)
        train_labels.append(pressure_level)
train_data = np.array(train_data)
train_labels = np.array(train_labels)
# print(train_data.shape)
# print(train_labels.shape)
# save the data to Regression/
np.save("Regression/train_data.npy", train_data)
np.save("Regression/train_labels.npy", train_labels)

# np.save("Regression/test_data.npy", train_data)
# np.save("Regression/test_labels.npy", train_labels)
print(aaa)
```
